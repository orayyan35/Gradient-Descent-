      To perform the 'best fit  line, we will assume

```python
plt.scatter(x, y)
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color='green')
plt.xlabel('area house ')
plt.ylabel('price house')
plt.title('Gradient Descent ')
plt.show()
```
